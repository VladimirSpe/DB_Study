```python
import pandas
import sqlite3
```

```python
data = pandas.read_csv("users (1).csv", delimiter='\t', encoding='koi8_r')
d = []
g = []
for i in data.values:
  if i[0] not in d:
    d.append(i[0])
    g.append((i[0].lower(), i[1], i[2]))


for i in g:
        connection = sqlite3.connect('test.db')
        cursor = connection.cursor()
        cursor.execute('''INSERT INTO USERs VALUES (?, ?, ?)''', i)
        connection.commit()
        connection.close()

```

```python
def create_db():
    connection = sqlite3.connect('test.db')
    cursor = connection.cursor()
    cursor.execute('''
    CREATE TABLE LOG (
    user_id TEXT NOT NULL,
    time TEXT NOT NULL,
    BET INTEGER,
    win INTEGER
    )
    ''')
    cursor.execute('''
    CREATE TABLE USERS (
    user_id TEXT PRIMARY KEY,
    email TEXT,
    geo TEXT
    )
    ''')
    connection.commit()
    connection.close()
create_db()
```

```python
def include_values_log():
    data = pandas.read_csv("log (1).csv", delimiter=',')
    user_id = []
    time = []
    bet = []
    win = []
    for i in data.values:
        if str(i[0]) != '#error':
            user_id.append(str(i[0]).split('-')[1][1:])
        if str(i[1]) != 'nan':
            time.append(str(i[1][1:]))
        if str(i[2]) != 'nan':
            bet.append(i[2])
        else:
            bet.append(0)
        if str(i[3]) != 'nan':
            win.append(i[3])
        else:
            win.append(0)

    for i in zip(user_id, time, bet, win):
        connection = sqlite3.connect('test.db')
        cursor = connection.cursor()
        cursor.execute('''INSERT INTO LOG VALUES (?, ?, ?, ?)''', i)
        connection.commit()
        connection.close()
include_values_log()
```

```python
#TASK 2
select = "select sum(win_per) / count(win_per) from (select user_id, (win * 100 /bet) as win_per from LOG where bet and win)"

conn = sqlite3.connect('test.db')
curr = conn.cursor()
res = curr.execute(select)
res.fetchall()
```

```python
#TASK 3
select = "select user_id, sum(win) - sum(bet) as balance from log group by user_id"

conn = sqlite3.connect('test.db')
curr = conn.cursor()
res = curr.execute(select)
res.fetchall()
```

```python
#TASK 4
select = """SELECT USERs.geo, (LOG.win - LOG.bet) from USERs
INNER JOIN LOG on LOG.user_id == USERS.user_id
ORDER BY (LOG.win - LOG.bet) DESC"""

conn = sqlite3.connect('test.db')
curr = conn.cursor()
res = curr.execute(select)
res.fetchall()
```

```python
#TASK 5
select = """SELECT DISTINCT USERs.geo, LOG.BET from USERS
INNER JOIN LOG on LOG.user_id = USERS.user_id
ORDER BY LOG.BET DESC"""

conn = sqlite3.connect('test.db')
curr = conn.cursor()
res = curr.execute(select)
res.fetchall()
```

```python
#TASK 6
select = """SELECT AVG(time_in - time_without) as time_delta FROM ((SELECT user_id as id, min(strftime('%s', LOG.time)) as time_without from LOG group by user_id)
INNER JOIN (SELECT user_id as id2, min(strftime('%s', LOG.time)) as time_in from LOG WHERE LOG.BET group by user_id) on id = id2)"""

conn = sqlite3.connect('test.db')
curr = conn.cursor()
res = curr.execute(select)
res.fetchall()
```
